```python
"""
示例4：构建多轮对话的模板

包含系统提示、对话历史和当前问题
"""
print("\n" + "="*70)
print("示例 4：多轮对话模板")
print("="*70)
```

```python
# 创建包含对话历史的模板
template = ChatPromptTemplate.from_messages([
    ("system", "你是一个{role}。{instruction}"),
    ("user", "{question1}"),
    ("assistant", "{answer1}"),
    ("user", "{question2}")
])
```

```python
messages = template.format_messages(
    role="Python 专家",
    instruction="回答要简洁、准确",
    question1="什么是列表？",
    answer1="列表是 Python 中的有序可变集合，用方括号 [] 表示。",
    question2="它和元组有什么区别？"  # 基于上下文的问题
)
```

```python
messages
```

```python
response = model.invoke(messages)
response
```

```python
"""
示例5：使用 MessagePromptTemplate 类

提供更细粒度的控制
"""
print("\n" + "="*70)
print("示例 5：MessagePromptTemplate 类（高级用法）")
print("="*70)
```

```python
system_template = SystemMessagePromptTemplate.from_template(
    "你是一个{profession}，你的特长是{specialty}"
)
```

```python
human_template = HumanMessagePromptTemplate.from_template(
    "关于{topic}, 我想知道{question}"
)
```

```python
chat_template = ChatPromptTemplate.from_messages([
    system_template,
    human_template
]
)
```

```python
# 使用模板
messages = chat_template.format_messages(
    profession="数据科学家",
    specialty="用数据讲故事",
    topic="数据可视化",
    question="如何选择合适的图表类型？"
)
messages
```

```python
response = model.invoke(messages)
response
```

```python
"""
示例6：部分变量 - 预填充某些变量

适用场景：
- 某些变量固定不变
- 需要创建模板变体
"""
print("\n" + "="*70)
print("示例 6：部分变量（Partial Variables）")
print("="*70)
```

```python
original_template = ChatPromptTemplate.from_messages([
    ("system", "你是一个{role}，你的目标用户是{audience}。"),
    ("user", "请{task}")
])
```

```python
original_template.input_variables
```

```python
# 部分填充：固定 role 和 audience
partially_filled = original_template.partial(
    role="科技博客作者",
    audience="程序员"
)
```

```python
partially_filled
```

```python
messages1 = partially_filled.format_messages(
    task="写一篇关于 Python 装饰器的文章开头"
)
messages1
```

```python
model.invoke(messages1)
```

```python
for chunk in model.stream(messages1):
    print(chunk.content, end="")
```

```python
"""
示例9：模板 + 模型的链式调用

LangChain Expression Language (LCEL)
"""
print("\n" + "="*70)
print("示例 9：LCEL 链式调用（预览）")
print("="*70)
```

```python
template = ChatPromptTemplate.from_messages([
    ("system","你是一个{role}"),
    ("user","{input}")
])
template
```

```python
chain = template | model
```

```python
chain.invoke({
    "role":"历史学家",
    "input":"请你解释一下安史之乱"
})
```
